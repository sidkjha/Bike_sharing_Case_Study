```python
df = pd.concat([df,df_dummies],axis=1)
df.head()
```

```python
#splitting data
x=df.drop(["cnt","instant"],axis=1)
x.head(2)
```

```python
y=df["cnt"]
y.head(2)
```

```python
#Splitting data into train test
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.30,random_state=100)
```

```python
#feature scaling
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
x_train[["yr","temp","hum","windspeed","casual","registered"]]=scaler.fit_transform(x_train[["yr","temp","hum","windspeed","casual","registered"]])
```

```python
x_train
```

```python
x_test[["yr","temp","hum","windspeed","casual","registered"]]=scaler.transform(x_test[["yr","temp","hum","windspeed","casual","registered"]])
```

```python
x_test
```

```python
from sklearn.ensemble import GradientBoostingRegressor
model = GradientBoostingRegressor()
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```
